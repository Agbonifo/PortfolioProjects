```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://www.worldometers.info/geography/largest-countries-in-the-world/"

page = requests.get(url)

soup = BeautifulSoup(page.text, "html")
```

```python
# Format with prettify

# soup.prettify()
```

```python
# Select table

table = soup.find("table")

# table
```

```python
 # Select table fields

table_head = soup.find("table")

table2_columns = table_head.find("tr")
table2_columns_formatted = [col.text.strip() for col in table2_columns]

# table2_columns_formatted
```

```python
# Remove empty strings from the list of formatted table fields 
table2_columns_formatted = [col for col in table2_columns_formatted if col]  

# table2_columns_formatted
```

```python
# Select table rows

table = soup.find("table")

table = table.findAll("tr")

row_data_list = []  
    
# Loop through the rows, skipping the header  
for row in table[1:]:  
    rows = row.find_all('td')  
    if rows:  
            # Scrap each relevant data point  
        row_data = [row.get_text(strip=True) for row in rows]
        row_data_list.append(row_data)  

# row_data_list 
```

```python
# Creating a dataframe of largest-countries-in-the-world and saving it

df = pd.DataFrame(row_data_list, columns=table2_columns_formatted)

df.to_excel("world_area.xlsx", index=False)

df
```
