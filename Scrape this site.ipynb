<b style="font-size: 1.5rem">Scraping all the pages data at once</b>

```python
# Define the base URL and the total number of pages  
base_url = 'https://www.scrapethissite.com/pages/forms/?page_num='  
total_pages = 24  

def scrape_page(page_num):  
    # Send an HTTP request to the page  
    response = requests.get(base_url + str(page_num))  
    
    # Check if the request was successful  
    if response.status_code != 200:  
        print(f"Failed to retrieve data from page {page_num}")  
        return []  
    
    # Parse the HTML content  
    soup = BeautifulSoup(response.content, 'html')  
    
    # Find the table and extract rows    
    table_rows = soup.select('table tr')  
    
    data_list = []  
    
    # Loop through the rows, skipping the header  
    for row in table_rows[1:]:  
        columns = row.find_all('td')  
        if columns:  
            # Scrap each relevant data point  
            data = [col.get_text(strip=True) for col in columns]  
            data_list.append(data)  

    return data_list  

# List to hold data from all pages  
all_data = []  

# Loop through all the pages and collect data  
for page in range(1, total_pages + 1):  
    # print(f"Scraping page {page}...")  
    page_data = scrape_page(page)  
    all_data.extend(page_data)  

# Convert the collected data into a DataFrame   
df_all = pd.DataFrame(all_data, columns=table_columns)  

# Save the DataFrame to a CSV file  
df_all.to_csv('hockey_stats.csv', index=False)  

# Display the final DataFrame  
df_all
```
